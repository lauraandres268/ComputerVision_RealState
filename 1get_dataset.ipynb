- Este df tiene algún problema respecto a las columnas de latitude y longitude
- Hacemos una normalización previa para que estén igual que los otros tres datasets

```python
import pandas as pd
import numpy as np

def normalize_coord(coord, is_latitude=True):

    try:
        num = float(coord)
        if num == 0:
            return num

        abs_num = abs(num)
        sign = -1 if num < 0 else 1

        # Determinar la posición del punto decimal basado en el orden de magnitud
        if is_latitude:
            # Latitud: esperamos valores ~40 (2 dígitos enteros)
            if abs_num > 999:
                return sign * (abs_num / 1000)
            elif abs_num > 99:
                return sign * (abs_num / 100)
            elif abs_num > 9:
                return sign * (abs_num / 10)
        else:
            # Longitud: esperamos valores ~-3.7 (1 dígito entero)
            if abs_num > 999:
                return sign * (abs_num / 1000)
            elif abs_num > 99:
                return sign * (abs_num / 100)
            elif abs_num > 9:
                return sign * (abs_num / 10)

        return num
    except:
        return np.nan
```

```python

# Aplicar a tu DataFrame
df['latitude'] = df['latitude'].apply(lambda x: normalize_coord(x, is_latitude=True))
df['longitude'] = df['longitude'].apply(lambda x: normalize_coord(x, is_latitude=False))
df['latitude'] = df['latitude'].apply(lambda x: normalize_coord(x, is_latitude=True))
df['longitude'] = df['longitude'].apply(lambda x: normalize_coord(x, is_latitude=False))
```

- creo columnas nuevas:

```python
import numpy as np

# Función para mapear exterior/interior
def map_exterior(desc):
    if pd.isna(desc):
        return np.nan
    desc = str(desc).lower()
    if 'exterior' in desc:
        return 1
    elif 'interior' in desc:
        return 0
    else:
        return np.nan

# Función para mapear ascensor
def map_ascensor(desc):
    if pd.isna(desc):
        return np.nan
    desc = str(desc).lower()
    if 'con ascensor' in desc:
        return 1
    elif 'sin ascensor' in desc:
        return 0
    else:
        return np.nan


df_final['exterior'] = df_final['floor_description'].apply(map_exterior)
df_final['ascensor'] = df_final['floor_description'].apply(map_ascensor)

print("Distribución de valores para 'exterior':")
print(df_final['exterior'].value_counts(dropna=False))
print("\nDistribución de valores para 'ascensor':")
print(df_final['ascensor'].value_counts(dropna=False))
```

### 2.3 Imputación

**imputacion con nlp en descripción**

```python
import re
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier

def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower() #minusculas
    text = re.sub(r'[^\w\s]', '', text)  # Elimina puntuación
    return text

df_final['cleaned_desc'] = df_final['property_description'].apply(clean_text)
```

```python
vectorizer = TfidfVectorizer(
    max_features=500,
    stop_words=['el', 'la', 'de', 'en', 'y', 'con']
)
X = vectorizer.fit_transform(df_final['cleaned_desc'])
```

- 1.Floor

```python
from sklearn.model_selection import train_test_split

train_data = df_final.dropna(subset=['floor'])
X_train = vectorizer.transform(train_data['cleaned_desc'])
y_train = train_data['floor'].astype(str)  # Convertir a string para clasificación

#clasificación (ajusta hiperparámetros si es necesario)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

#valores faltantes - busco filas con na y aplico solo a ella
null_indices = np.where(df_final['floor'].isna())[0]

if len(null_indices) > 0:
    X_null = X[null_indices]  #solo filas nulas
    df_final.loc[df_final.index[null_indices], 'floor'] = model.predict(X_null)

print(f"Valores imputados: {df_final['floor'].isna().sum()} nulos restantes")
print("Distribución de valores:\n", df_final['floor'].value_counts(dropna=False))
```

**imputacion exterior y ascensor (binarias)**

```python

import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

def impute_binary_column(col_name, positive_words, negative_words):
    y = df_final[col_name].copy()

    mask_pos = df_final['property_description'].str.contains(
        '|'.join([re.escape(word) for word in positive_words]), na=False, case=False
    )
    mask_neg = df_final['property_description'].str.contains(
        '|'.join([re.escape(word) for word in negative_words]), na=False, case=False
    )

    y = np.where(
        df_final[col_name].notna(),
        df_final[col_name],
        np.where(mask_pos, 1,
                 np.where(mask_neg, 0, np.nan))
    )

    train_mask = df_final[col_name].notna()
    train_indices = np.where(train_mask)[0]

    null_mask = df_final[col_name].isna() & np.isnan(y)
    null_indices = np.where(null_mask)[0]

    if len(train_indices) > 0:
        print("aplicando classifier")
        model = RandomForestClassifier(n_estimators=100, random_state=42)
        model.fit(X[train_indices], y[train_indices].astype(int))

        if len(null_indices) > 0:
            df_final.loc[df_final.index[null_indices], col_name] = model.predict(X[null_indices])

    print(f"\n=== Reporte para {col_name} ===")
    if len(train_indices) > 0:
        print(classification_report(y[train_indices], model.predict(X[train_indices])))
    else:
        print("No hay datos etiquetados para entrenar el modelo.")


#palabras clave para cada variable para buscarlas lo primero
exterior_words = {
    'positive': ['exterior', 'fachada', 'balcón', 'terraza'],
    'negative': ['interior', 'patio interior']
}

ascensor_words = {
    'positive': ['ascensor', 'elevador'],
    'negative': ['sin ascensor', 'no ascensor']
}

# 5. Ejecutar imputación
impute_binary_column('exterior', exterior_words['positive'], exterior_words['negative'])
impute_binary_column('ascensor', ascensor_words['positive'], ascensor_words['negative'])

# 6. Resultados finales
print("\nDistribución de valores:")
print("exterior:\n", df_final['exterior'].value_counts(dropna=False))
print("\nascensor:\n", df_final['ascensor'].value_counts(dropna=False))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


# Configuración de estilo
MODERN_PALETTE = ['#4C72B0', '#DD8452', '#55A868', '#C44E52']
sns.set_style("whitegrid", {'grid.color': '#f0f0f0'})
sns.set_palette(MODERN_PALETTE)

# Columnas a visualizar (ajusta según tu DataFrame)
columns_to_plot = [
    'bathroom_count',  # Número de baños
    'floor',           # Número de piso
    'property_type',   # Tipo de propiedad
    'energy_certificate',  # Certificado energético
    'exterior',        # ¿Es exterior?
    'ascensor'         # ¿Tiene ascensor?
]

# Mapeo de nombres para títulos (opcional)
title_names = {
    'bathroom_count': 'Número de Baños',
    'floor': 'Planta/Piso',
    'property_type': 'Tipo de Propiedad',
    'energy_certificate': 'Certificado Energético',
    'exterior': 'Propiedades Exteriores',
    'ascensor': 'Presencia de Ascensor'
}

# Bucle para generar los gráficos
for column in columns_to_plot:
    # Verificar si la columna existe en los datos
    if column not in dfff.columns:
        print(f"\n⚠️ Columna '{column}' no encontrada. Saltando...")
        continue

    print(f"\nGenerando gráfico para: {column}")

    # Configuración específica por tipo de columna
    if column in ['bathroom_count', 'floor']:
        # Para variables numéricas discretas
        data = dfff[column].dropna()
        order = sorted(data.unique()) if data.dtype.kind in 'iuf' else None
    else:
        # Para variables categóricas
        data = dfff[column].dropna()
        order = data.value_counts().index[:10]  # Mostrar hasta 10 categorías

    # Crear tabla cruzada con neighborhood
    try:
        cross_tab = pd.crosstab(data, dfff['neighborhood'])
    except Exception as e:
        print(f"Error al procesar {column}: {str(e)}")
        continue

    # Configurar figura
    plt.figure(figsize=(12, 6))

    # Gráfico de barras apiladas
    ax = cross_tab.plot(
        kind='bar',
        stacked=True,
        color=MODERN_PALETTE,
        width=0.85,
        edgecolor='white',
        linewidth=0.5
    )

    # Títulos y etiquetas
    title = title_names.get(column, column.replace('_', ' ').title())
    plt.title(f'Distribución de {title} por Barrio', fontsize=12, pad=15)
    plt.xlabel(title, fontsize=10)
    plt.ylabel('Número de Propiedades', fontsize=10)
    plt.xticks(rotation=45 if len(cross_tab) > 5 else 0, fontsize=9)
    plt.yticks(fontsize=9)

    # Leyenda
    barrio_labels = {
        1: 'Chamberí',
        2: 'Centro',
        3: 'Arganzuela',
        4: 'Retiro'
    }

    ax.legend(
        [barrio_labels[int(float(label))] for label in cross_tab.columns],
        title='Barrio',
        loc='upper right',
        frameon=True,
        framealpha=0.9,
        fontsize=9,
        title_fontsize=10
    )

    # Añadir etiquetas de valor
    for rect in ax.patches:
        height = rect.get_height()
        if height > 0:
            ax.text(
                rect.get_x() + rect.get_width()/2,
                rect.get_y() + height/2,
                f'{int(height)}',
                ha='center',
                va='center',
                color='white',
                fontweight='bold',
                fontsize=8,
                alpha=0.9
            )

    plt.tight_layout()
    plt.show()
```

```python
# 1. Preparación de datos (asegurar mayúsculas)
dfff['energy_certificate'] = dfff['energy_certificate'].str.upper()
cert_order = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

# 2. Filtrar solo certificados válidos
valid_certs = dfff[dfff['energy_certificate'].isin(cert_order)].copy()

# 3. Mapeo de nombres de barrios
valid_certs['neighborhood_name'] = valid_certs['neighborhood'].map({
    1: 'Chamberí',
    2: 'Centro',
    3: 'Arganzuela',
    4: 'Retiro'
})

# 4. Configuración del gráfico
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid", {'grid.color': '#f0f0f0'})

# 5. Crear tabla cruzada
cross_tab = pd.crosstab(valid_certs['energy_certificate'], valid_certs['neighborhood_name'])

# 6. Ordenar por certificado energético
cross_tab = cross_tab.loc[cert_order]

# 7. Gráfico de barras apiladas
ax = cross_tab.plot(
    kind='barh',  # Horizontal
    stacked=True,
    color=['#4C72B0', '#DD8452', '#55A868', '#C44E52'],  # Colores por barrio
    width=0.85,
    edgecolor='white',
    linewidth=0.5
)

# 8. Personalización
plt.title('Certificados Energéticos por Barrio', fontsize=12, pad=15)
plt.xlabel('Número de Propiedades', fontsize=10)
plt.ylabel('Certificado Energético', fontsize=10)
plt.yticks(fontsize=9)
plt.xticks(fontsize=9)

# 9. Leyenda dentro del gráfico
ax.legend(
    title='Barrio',
    loc='center left',
    bbox_to_anchor=(1, 0.5),
    frameon=True,
    framealpha=0.9,
    fontsize=9,
    title_fontsize=10
)

# 10. Añadir etiquetas de valor
for rect in ax.patches:
    width = rect.get_width()
    if width > 0:
        ax.text(
            rect.get_x() + width/2,
            rect.get_y() + rect.get_height()/2,
            f'{int(width)}',
            ha='center',
            va='center',
            color='white',
            fontweight='bold',
            fontsize=8
        )

plt.tight_layout()
plt.show()
```

```python
# Función para calcular moda con porcentaje
def moda_con_porcentaje(serie):
    counts = serie.value_counts(normalize=True)
    moda = counts.index[0]
    porcentaje = round(counts.iloc[0] * 100, 1)
    return pd.Series([moda, porcentaje], index=['moda', 'porcentaje'])

# Cálculo de todas las métricas con porcentajes
metrics = {
    # Certificado energético ya tiene %
    'certificado': (dfff.groupby('neighborhood_name')['energy_certificate']
                   .value_counts(normalize=True)
                   .groupby(level=0).head(1)
                   .mul(100).round(1)),

    # Baños con %
    'baños': dfff.groupby('neighborhood_name')['bathroom_count'].apply(moda_con_porcentaje),

    # Planta con %
    'planta': dfff.groupby('neighborhood_name')['floor'].apply(moda_con_porcentaje),

    # Dormitorios con %
    'dormitorios': dfff.groupby('neighborhood_name')['bedroom_count'].apply(moda_con_porcentaje),

    # Precios (sin %)
    'precio': dfff.groupby('neighborhood_name')['property_price'].agg(['mean', 'median']).round(),

    # Tipo propiedad ya tiene %
    'tipo_propiedad': (dfff.groupby('neighborhood_name')['property_type']
                      .value_counts(normalize=True)
                      .groupby(level=0).head(1)
                      .mul(100).round(1))
}

# Mostrar resultados mejor formateados
print("=== DATOS COMPLETOS PARA TABLAS LATEX ===")
for nombre, datos in metrics.items():
    print(f"\n🔷 {nombre.upper()}:")
    if isinstance(datos, pd.Series):
        print(datos.to_string())
    else:
        print(datos.reset_index().to_string(index=False))
```

```python
# Calcular estadísticas de precios
price_stats = dfff.groupby('neighborhood_name')['property_price'].agg(['mean', 'min', 'max']).round().astype(int)

# Calcular moda y porcentaje de tipo de propiedad
property_stats = (dfff.groupby('neighborhood_name')['property_type']
                  .apply(lambda x: x.mode()[0] + f" ({round(x.value_counts(normalize=True).iloc[0]*100, 1)}\%)"))

# Combinar resultados
final_stats = pd.concat([price_stats, property_stats], axis=1)
final_stats.columns = ['Promedio', 'Mínimo', 'Máximo', 'Tipo Propiedad']
print(final_stats.to_markdown())
```
